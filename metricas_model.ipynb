```python
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# Ignorar avisos
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/americanas-tech/b2w-reviews01/main/B2W-Reviews01.csv")
df.head()
```

```python
from papalinguas.core import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()

def get_sentiment(text, limiar=0.0):

    scores = analyzer.polarity_scores(text)

    sentiment = 0 if scores > limiar else 1

    return sentiment




# apply get_sentiment function

df['sentiment'] = df['review_text_lemmatized'].apply(lambda x: get_sentiment(str(x)))

df
```

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(df['target'], df['sentiment']))
```

```python
from sklearn.metrics import classification_report

print(classification_report(df['target'], df['sentiment']))
```
